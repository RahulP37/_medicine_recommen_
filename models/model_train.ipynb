```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("noorsaeed/medicine-recommendation-system-dataset")

print("Path to dataset files:", path)
```

```python
import os
os.listdir(path)
```

```python
import pandas as pd
df=pd.read_csv(os.path.join(path,"Training.csv"))
df.head()
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.shape
```

```python
df['prognosis'].value_counts()
```

```python
len(df['prognosis'].unique())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X=df.drop('prognosis',axis=1)
y=df['prognosis']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=24)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
encoder=LabelEncoder()
```

```python
encoder.fit(y)
```

```python
y=encoder.transform(y)
y
```

```python
y_train=encoder.transform(y_train)
y_test=encoder.transform(y_test)
```

```python
import numpy as np
disease_dict = {}
for label in np.unique(y):  # y is the encoded array
    disease_name = encoder.inverse_transform([label])[0]
    disease_dict[label] = disease_name
disease_dict
```

```python
symp_dict={}
for i,symp in enumerate(df.columns):
  if symp=='prognosis':
    break
  symp_dict[symp.strip()]=i
symp_dict
```

```python
x_train.columns
```

# training top models

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

models = {
    "SVC_linear": (SVC(kernel='linear'), {}),
    "SVC_rbf": (SVC(kernel='rbf'), {'C': [0.1, 1, 10]}),
    "SVC_sigmoid": (SVC(kernel='sigmoid'), {'C': [0.1, 1, 10]}),
    "RandomForest": (RandomForestClassifier(), {'n_estimators': [10, 50, 100]}),
    "DecisionTree": (DecisionTreeClassifier(), {'max_depth': [None, 5, 10]}),
    "KNN": (KNeighborsClassifier(), {'n_neighbors': [3, 5, 7]})
}

best_models = {}

for name, (model, params) in models.items():
    grid = GridSearchCV(model, params, cv=5, scoring='accuracy')
    grid.fit(x_train, y_train)

    print(f"Best params for {name}: {grid.best_params_}")
    print(f"Best CV score for {name}: {grid.best_score_:.4f}")

    best_models[name] = grid.best_estimator_

    acc = grid.score(x_test, y_test)
    print(f"Accuracy score on test set for {name}: {acc:.4f}")
    print("-" * 100)
```

```python
svc = SVC(kernel='linear', probability=True)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
accuracy_score(y_test,y_pred)
```
